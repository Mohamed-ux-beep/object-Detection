**1. Reading the data**

```python
from google.colab import files
import zipfile
import os

uploaded = files.upload()

for fname in uploaded.keys():
  if fname.endswith('.zip'):
    with zipfile.ZipFile(fname, 'r') as zip_ref:
      zip_ref.extractall()
```

**2. Training YOLO model**

```python
from ultralytics import YOLO
import os

# load YOLO
model = YOLO("yolov8n.pt")

# train the model
model.train(data="/content/dataset-ki/data.yaml", epochs=50, imgsz=640)
```

```python
trained_model = YOLO("/content/runs/detect/train/weights/best.pt")
result = trained_model.predict(source="/content/dataset-ki/images/train/car_01.jpeg", conf=0.3, save=True)
```
